### Forward Propogation



<h1><b>Backpropagation</b></h1>

Backpropagation algorithm is an algorithm that can be used to train many kinds of neural network architecture. Backpropagation is named so as there is a backward propagation of errors in order to adapt the weights or parameters in the neural network.



Before diving deep into backpropagation let's understand concept of **chain rule** and **local gradient**.

In calculus, the chain rule is a formula to compute the derivative of a composite function. Let's assume we have an input $x$ and $y=f(x)$ and $z=g(y)$.

$\frac{\partial{z}}{\partial{x}}$ is change in value of $z$ with change in $x$ value. Since $z$ depend on $y$ and $y$ depend on $x$, $z$ is linked with $x$ through an intermediate variable $y$. In this case, the chain rule states that:$$\frac{\partial{z}}{\partial{x}}=\frac{\partial{z}}{\partial{y}}.\frac{\partial{y}}{\partial{x}}$$

Suppose, we have $x$ and $y$ as input which is feed to $z$, where $z=f(x,y)$ and $z$ is finally feed to $L$,where $L=g(z)$.
