# Chi-square Test

```python
import scipy.stats as stats
```

```python
contengency_table = pd.crosstab(titanic_df["Pclass"], titanic_df["SibSp"])
contengency_table
```

```python
chi2_score, p_value = stats.chi2_contingency(contengency_table)
```
