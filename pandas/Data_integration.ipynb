# Data integration


## 1. Concatenation

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(
    {
        "A" :[1,2,3],
        "B" : [3,4,5],
        "C" : [6,9,0]
    }
)
df2 = pd.DataFrame(
    {
        "A" :[23,43,55],
        "B" : [31,42,53],
        "C" : [67,49,90]
    }
)
```

```python
#pd.concat()
row_concat = pd.concat([df1, df2], axis = 0)
row_concat 
```

```python
col_concat = pd.concat([df1, df2], axis = 1)
col_concat 
```

## 2. Merge

pd.merge()

```python
df1 = pd.DataFrame(
    {"key" : ["A","B","C","D"], "ins_val" : [23,34,56,11], "acc": [4,6,0,1]}
)
df2 = pd.DataFrame(
    {"key" : ["E","B","C","Z"], "hel_ratio" : [22,90,78,45], "acc": [5,7,9,1]}
)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how = "inner", on = "key")
```

```python
pd.merge(df1, df2, how = "outer", on = "key", suffixes = ("_df1", "_df2"))
```

```python
pd.merge(df1, df2, how = "left", on = "key", suffixes = ("_df1", "_df2"))
```

```python
pd.merge(df1, df2, how = "right", on = "key", suffixes = ("_df1", "_df2"))
```

```python
# merge withdf
df1.merge(df2, how ="right", on = "key")
```

## 3. Joining

```python
df1 = pd.DataFrame(
    {"key" : ["A","B","C","D"], "ins_val" : [23,34,56,11]}
)
df2 = pd.DataFrame(
    {"key" : ["E","B","C","Z"], "hel_ratio" : [22,90,78,45]}
)
```

```python
df1 = df1.set_index("key")
df2 = df2.set_index("key")
```

```python
df1.join(df2, how= "inner")
```

```python
df1.join(df2, how= "outer")
```

## 4. Stacking


dataframe.stack()

```python
df1 = pd.DataFrame(
    {
        "A" :[1,2,3],
        "B" : [3,4,5],
        "C" : [6,9,0]
    },
    index = ["X","Y","Z"]
)
df2 = pd.DataFrame(
    {
        "A" :[23,43,55],
        "B" : [31,42,53],
        "C" : [67,49,90]
    },
    index = ["X","Y","Z"]
)
```

```python
stacked =df1.stack()
stacked
```

```python
stacked.unstack()
```
