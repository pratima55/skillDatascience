## 1. Bar graph

```python
titanic_df["Survived"].value_counts().plot(
    kind = "bar", legend = True, figsize = (4,4),
    title = "count of survived Passengers",
    grid = True,
    fontsize = 10,
    colormap = "Dark2"
)
```

```python
group_by_gender= titanic_df[["Sex", "Survived"]].groupby(by = "Sex")
group_by_gender.value_counts().plot(
    kind = "bar",
    colormap = "Dark2",
    grid = True
)
```

```python
 group_by_embarked= titanic_df[[ "Age", "Embarked"]].groupby(by = "Embarked")
group_by_embarked.mean().plot(
    kind = "bar",
    grid = True,
    colormap = "Dark2"
)
```

```python
ax= titanic_df[[ "Age", "Embarked"]].groupby(by = "Embarked")
ax
```
